```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(5)
np.random.seed(0)
y=np.random.randint(0,100,size=(5,))
plt.scatter(x,y,c=set("rgbmy"))
plt.xlabel("5 days ")
plt.ylabel("percentege")
plt.title("Scattar Plot")
plt.show()
```

```python
plt.scatter(x,y,c=set("rgbmy"),s=(333,222,111,232,453),marker="+")
plt.show()
```

```python
plt.scatter(x,y,c=set("rgbmy"),alpha=0.4,s=(333,222,111,232,453),marker="o")
plt.show()
```

```python
plt.scatter(x,y,c=set("rgbmy"),marker="+",s=(333,222,111,232,453),edgecolor='r',linewidth=5)
plt.show()
```

```python
color=np.linspace(0,100,5)
plt.scatter(x,y,c=color,cmap="BrBG")
plt.colorbar()
```

```python
# drawing two plots on one scatter plot 
x_=np.arange(5)
y_=np.random.randint(0,100,size=(5,))
z_=np.linspace(0,80,5)
plt.subplots()
plt.scatter(x_,y_,c='r')
plt.scatter(x_,z_,c='y')
plt.show()
```

```python
x_=np.arange(5)
y_=np.random.randint(0,100,size=(5,))
z_=np.linspace(0,80,5)
plt.subplots()
plt.scatter(x_,y_,c=color,cmap='viridis',label="Plot 1")
plt.scatter(x_,z_,c=color,cmap='viridis',label="Plot 2")
plt.colorbar()
plt.legend()
plt.show()
```
