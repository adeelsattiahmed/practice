```python
import numpy as np 
import matplotlib.pyplot as plt
x=['python','C++','JavaScript','HTML']
np.random.seed(0)
y=np.random.randint(10,100,size=(4,))
```

```python
y
```

```python
plt.subplots()
plt.bar(x,y,width=0.4,color=set("rgbm"))
plt.show()
```

```python
plt.subplots()
plt.bar(x,y,color=set("rgbm"),label='Popularity')
plt.xlabel('Languages',fontsize=14)
plt.ylabel('Percentage',fontsize=14)
plt.title('Comparison',fontsize=14)
plt.legend()
plt.show()
```

```python
z=[20,30,40,50]
width=0.4
plt.subplots()
plt.bar(x,y,width,color=set("r"),label='Popularity')
plt.bar(x,z,width,color=set("m"),label='Popularity2')
plt.xlabel('Languages',fontsize=14)
plt.ylabel('Percentage',fontsize=14)
plt.title('Comparison',fontsize=14)
plt.legend()
plt.show()
```

```python
p=np.arange(len(x))
```

```python
p1=[j+width for j in p]
```

```python
plt.subplots()
plt.bar(p,y,width,color=set("r"),label='Popularity1')
plt.bar(p1,z,width,color=set("m"),label='Popularity2')
plt.xlabel('Languages',fontsize=14)
plt.ylabel('Percentage',fontsize=14)
plt.title('Comparison',fontsize=14)
plt.xticks(p+width,x)
plt.legend()
plt.show()
```

```python
plt.subplots()
plt.bar(p,y,width,color=set("r"),label='Popularity1')
plt.bar(p1,z,width,color=set("m"),label='Popularity2')
plt.xlabel('Languages',fontsize=14)
plt.ylabel('Percentage',fontsize=14)
plt.title('Comparison',fontsize=14)
plt.xticks(p+width/2,x)
plt.legend()
plt.show()
```

```python
plt.subplots()
plt.barh(p,y,width,color=set("r"),label='Popularity1')
plt.barh(p1,z,width,color=set("m"),label='Popularity2')
plt.xlabel('Languages',fontsize=14)
plt.ylabel('Percentage',fontsize=14)
plt.title('Comparison',fontsize=14)
plt.yticks(p+width/2,x)
plt.legend()
plt.show()
```
