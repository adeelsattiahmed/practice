```python
import numpy as np 
import matplotlib.pyplot as plt 
x=np.arange(5)
np.random.seed(0)
y=np.random.randint(1,100,size=(5,))
w=2*y
z=3*y
u=[20,30,350,200,400]
fig,((a,b),(c,d),(e,f))=plt.subplots(nrows=3,ncols=2,figsize=(30,30))
a.plot(x,y,'ro-',label='Line plot')
a.set(xlabel="X_axis",ylabel="Y_axis",title="Plot 1")
a.grid()
a.legend()
b.bar(x,w,color='g',label='Bar plot')
b.set(xlabel="X_axis",ylabel="Y_axis",title="Plot 2")
b.legend()
b.grid()
c.scatter(x,z,c=set('rgbym'),s=[200,300,400,230,349],label='scatter plot',marker="o")
c.set(xlabel="X_axis",ylabel="Y_axis",title="Plot 3")
c.legend()
c.grid()
d.hist(u,color='r',label='Histogram Plot',rwidth=0.5)
d.set(title="Plot 4")
d.axvline(300,color='g',label='line')
d.grid()
d.legend()
w=[10,20,30,40,50]
z=['C','C++','Python','Javascript','Artficial Intelligence']
c=[0,0,0,0,0.1]
e.pie(w,labels=z,explode=c,autopct="%0.1f%%",radius=1.5,labeldistance=1.1,center=(0,1),textprops={'fontsize':15})
e.pie([1],colors='w',labeldistance=1.1,center=(0,1))
e.legend(loc=3)
e.grid()
```
